```python
def slicegrayscalemap(graymap, max_level, index):
    b = np.full((graymap.shape[0], graymap.shape[1]), 0)
    c = np.full((graymap.shape[0], graymap.shape[1]), 1)
    a = np.where(graymap > 255 * index / max_level, c, b)
    return a
```

```python
for i in range(6):
    slice1 = slicegrayscalemap(gray, 6, i)
    plt.show() 
    plt.imshow(slice1/256, cmap='gray')
```
